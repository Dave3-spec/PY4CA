```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
raw_file = np.genfromtxt('Accra_1983_2017_dRR_gapless.txt')
```

```python
data = pd.DataFrame(raw_file, columns=['Year','Month','Day','Val'])
data['Year']=data['Year'].astype(int)
data['Month']=data['Month'].astype(int)
data['Day']=data['Day'].astype(int)
```

```python
data_m = data.groupby(['Year','Month'])['Val'].sum().reset_index()
data_m.head()
```

```python
year = data_m['Year']
precip = data_m['Val']
value=[]

for yy in np.arange(1983,2017):
    pp=precip[year==yy]
    R=np.sum(pp)
    o=R/12
    B=pp-o
    c=np.abs(B)
    m=np.sum(c)
    SI = m/R
    value.append((yy,SI))
    
   # print(c)
    
```

```python
print(value)
```

```python
T=pd.DataFrame(value,columns=['Year','SI'])
T=T.set_index('Year')
```

```python
print(T)
```

```python
T.plot(marker='o', linestyle='-')
plt.title('SI for Abetifi-Ghana')
```

```python
sb.distplot(T['SI'], color='k')
plt.title('Seasonality Distribution')
```
